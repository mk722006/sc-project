```python
import csv
import json
import psycopg2

def create_table(cursor):
    # Create a SQL query to create the table
    create_table_query = '''
    CREATE TABLE IF NOT EXISTS your_table_name (
        DT_RowId VARCHAR(255),
        first_name VARCHAR(255),
        last_name VARCHAR(255),
        phone VARCHAR(255),
        site VARCHAR(255),
        site_name VARCHAR(255)
    );
    '''
    cursor.execute(create_table_query)

def insert_data(cursor, data):
    # Extract user data
    user_data = data.get('users', {})
    
    # Extract site data
    site_data = data.get('sites', {})

    # Create a SQL query to insert data into the table
    insert_query = '''
    INSERT INTO your_table_name (DT_RowId, first_name, last_name, phone, site, site_name)
    VALUES (%s, %s, %s, %s, %s, %s);
    '''
    cursor.execute(insert_query, (
        data.get('DT_RowId', ''),
        user_data.get('first_name', ''),
        user_data.get('last_name', ''),
        user_data.get('phone', ''),
        user_data.get('site', ''),
        site_data.get('name', ''),
    ))

def json_to_csv(json_file_path, db_config):
    with open(json_file_path, 'r') as json_file:
        data = json.load(json_file)

        # Connect to PostgreSQL
        connection = psycopg2.connect(**db_config)
        cursor = connection.cursor()

        try:
            # Create table if it doesn't exist
            create_table(cursor)

            # Insert data into the table
            insert_data(cursor, data)

            # Commit the changes
            connection.commit()
            print("Data successfully inserted into PostgreSQL!")

        except Exception as e:
            print(f"Error: {e}")

        finally:
            # Close the cursor and connection
            cursor.close()
            connection.close()

if __name__ == "__main__":
    #  path to your JSON file
    json_file_path = 'data.json'

    # Replace 'your_database_name', 'your_username', and 'your_password' with your PostgreSQL connection details
    db_config = {
        'host': 'localhost',
        'database': 'postgres',
        'user': 'postgres',
        'password': 'root',
    }   

    json_to_csv(json_file_path, db_config)
```
